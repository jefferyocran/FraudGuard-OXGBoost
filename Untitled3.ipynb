```python
!pip install -U xgboost shap -q

import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, average_precision_score

SEED = 42
np.random.seed(SEED)
print("Environment ready. Seed =", SEED)
```

```python
# Load UCI
url = "https://raw.githubusercontent.com/justmarkham/pycon-2016-tutorial/master/data/sms.tsv"
uci = pd.read_csv(url, sep='\t', header=None, names=['label', 'message'])
uci['label'] = uci['label'].map({'ham': 0, 'spam': 1})

# Load your field data (upload ghana_momo_sms.csv to this notebook too)
field = pd.read_csv('ghana_momo_sms.csv')[['label', 'message']]

# Combine
df = pd.concat([uci, field], ignore_index=True)

# Same TF-IDF and same split as baseline
vectorizer = TfidfVectorizer(max_features=1000, ngram_range=(1,2))
X = vectorizer.fit_transform(df['message'])
y = df['label']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=SEED
)
print("Data ready. Train:", X_train.shape[0], "Test:", X_test.shape[0])
```

```python
def focal_loss_objective(y_pred, dtrain):
    """
    Custom focal loss for XGBoost.
    Down-weights easy examples, focuses on hard ones.
    gamma = focusing strength, alpha = class balance weight.
    """
    gamma = 2.0
    alpha = 0.25
    y_true = dtrain.get_label()

    # Convert raw predictions to probabilities
    p = 1.0 / (1.0 + np.exp(-y_pred))

    # p_t: how confident the model is on the CORRECT class
    p_t = y_true * p + (1 - y_true) * (1 - p)
    alpha_t = y_true * alpha + (1 - y_true) * (1 - alpha)

    # The focal weight: small for easy examples, large for hard ones
    focal_weight = alpha_t * np.power(1 - p_t, gamma)

    # Gradient and hessian (what XGBoost needs to learn)
    grad = focal_weight * (p - y_true)
    hess = focal_weight * p * (1 - p) * (gamma * (1 - p_t) + 1)

    return grad, hess

print("Focal loss objective defined. This is the core of O-XGBoost.")
```

```python
# Convert data to XGBoost's internal format
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

# Same hyperparameters as baseline — ONLY the objective changes
params = {
    'max_depth': 6,
    'eta': 0.1,              # learning rate
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'seed': SEED
}

# Train using YOUR custom focal loss
o_xgboost = xgb.train(
    params,
    dtrain,
    num_boost_round=200,
    obj=focal_loss_objective   # <-- this line makes it O-XGBoost
)
print("O-XGBoost trained with focal loss.")
```

```python
# Get predictions (these come out as raw scores, convert to probabilities)
raw_preds = o_xgboost.predict(dtest)
probs = 1.0 / (1.0 + np.exp(-raw_preds))
preds = (probs >= 0.5).astype(int)

print("O-XGBoost (Focal Loss)")
print("  Accuracy: ", round(accuracy_score(y_test, preds), 4))
print("  Macro F1: ", round(f1_score(y_test, preds, average='macro'), 4))
print("  AUC-ROC:  ", round(roc_auc_score(y_test, probs), 4))
print("  AUC-PR:   ", round(average_precision_score(y_test, probs), 4))
```
